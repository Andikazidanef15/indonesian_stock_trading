# Investment Multiple and CAGR
Two alternative reward metrics that are more intuitive and easier to interpret

Investment multiple: Ending Value of 1 Dollar Invested

$$
\text{Multiple} = \text{Ending Value}/\text{Initial Investment}
$$

```python
msft
```

```python
multiple = (msft.Price.iloc[-1]/msft.Price.iloc[0])
multiple
```

Price increase (in %)

```python
(multiple - 1) * 100
```

Drawback of Investment Multiple: Doesn't take into account investment period. Meaningful only in conjunction with Investment Period

Compound Annual Growth Rate (CAGR): The constant (annual rate of return) that would be required for an investment to grow from its beginning to its ending balance, assuming the profits were reinvested at the end of each year of the investment lifespan

```python
start = msft.index[0]
end = msft.index[-1]
```

```python
td = end - start
td
```

```python
td_years = td.days / 365.25
td_years
```

```python
cagr = multiple ** (1 / td_years) - 1
cagr
```

```python
multiple_from_cagr = (1 + cagr) ** td_years
multiple_from_cagr
```

We have CAGR around 26%

CAGR can be used to compare investments with different investment horizons

## Discrete Compounding

Annual compounding -> Interests accrue once a year, at the end of the year

Your saving bank offers an interest rate of 8% p.a. (stated rate) with annual compounding on your savings (USD 100). Calculate the value of your savings account after one year and the corresponding effective annual interest rate

-> Interest are calculated and added to your saving accounts once at the end of each year

```python
PV = 100
r = 0.08
n = 1
```

```python
100 * 1.08
```

```python
FV = PV * (1 + r)**n
FV
```

```python
effective_annual_rate = (FV/PV)**(1/n) - 1
effective_annual_rate
```

Quarterly Compounding -> Interests accrue once a quarter at the end of quarter

Your saving bank offers an interest rate of 8% p.a. (stated rate) with quarterly compounding on your savings (USD 100). Calculate the value of your savings account after one year and the corresponding effective annual interest rate

```python
PV = 100
r = 0.08
n = 1
m = 4
FV = PV * (1 + r/m)**(n*m)
FV
```

Quarterly compounding is favourable as we earn compound interest

Monthly Compounding -> Interests accrue once a month at the end of month

## Continuous Compounding

Your saving bank offers an interest rate of 8% p.a. (stated rate) with continuous compounding on your savings (USD 100). Calculate the value of your savings account after one year and the corresponding effective annual interest rate

```python
PV = 100
r = 0.08
n = 1
m = 100000

# Using Discrete Compounding
FV = PV * (1 + r/m)**(n*m)
FV
```

```python
# Alternative method using exponential
FV = PV * np.exp(n * r)
FV
```

```python
# Using discrete approach
effective_annual_rate = (FV/PV)**(1/n) - 1
effective_annual_rate
```

```python
# Using continuous approach
effective_annual_rate = np.exp(r) - 1
effective_annual_rate
```

Assume that we can only observe FV and PV, how to calculate the interet rate

```python
# Using continuous approach #2
effective_interest_rate = np.log(FV/PV)
effective_interest_rate
```

```python
r = np.log(effective_annual_rate + 1)
r
```

# Normality of Financial Returns

```python
msft
```

```python
msft.describe()
```

```python
msft.log_ret.plot(kind="hist", figsize=(15, 8), bins=100, fontsize=13, density=True)
plt.xlabel("Daily Returns")
plt.ylabel("Frequency")
plt.title("Frequency Distribution of Returns")
plt.show()
```

Do MSFT Returns follows normal distribution?

A normally distributed random variable can be fully described by its:
* mean
* standard deviation

Higher central moments are zero:
* Skew = 0 (measures symmetries around mean)
* (Excess) Kurtosis = 0 (Positive excess kurtosis -> more observations in tail)

```python
mu = msft.log_ret.mean()
mu
```

```python
sigma = msft.log_ret.std()
sigma
```

```python
import scipy.stats as stats
```

```python
stats.skew(msft.log_ret.dropna()) # in a normal distribution -> skew == 0
```

```python
stats.kurtosis(msft.log_ret.dropna(), fisher=True) # in a normal distribution -> Kurtosis (Fisher) == 0
```

-> MSFT returns exhibit "Fat Tails" (extreme positive/negative outcomes)

```python
x = np.linspace(msft.log_ret.min(), msft.log_ret.max(), 10000)
x
```

```python
y = stats.norm.pdf(x, mu, sigma) # Creating y values for a normal distribution with mean mu, sigma standard deviation
y
```

```python
plt.figure(figsize=(10, 5))
plt.hist(msft.log_ret, bins=500, density=True, label="Frequency Distribution of Daily Returns MSFT")
plt.plot(x, y, label="Normal Distribution")
plt.title("Normal Distribution")
plt.xlabel("Daily Returns")
plt.ylabel("pdf")
plt.legend()
plt.show()
```

Testing the normality of MSFT returns based on the sample: Hypothesis test with H0 hypothesis: MSFT Returns follow a normal distribution

```python
z_stat, p_value = stats.normaltest(msft.log_ret.dropna())
```

```python
z_stat # high value -> reject h0
```

```python
p_value # value lower than 0.05 -> reject h0
```

MSFT Returns does not follow normal distribution, it has fat tails due to extreme events/outcomes -> The standard deviation of returns underestimate true risk

# Resampling/Smoothing

```python
monthly = msft.Price.resample("M").last()
monthly
```

```python
monthly.plot(figsize=(10, 5), fontsize=13)
plt.legend()
plt.show()
```

How will the Mean-Variance analysis change with smoothed data?

```python
freqs = ["A", "Q", "M", "W-Fri", "D"]
periods = [1, 4, 12, 52, 252]
ann_mean = []
ann_std = []

for i in range(len(freqs)):
    resampled = msft.Price.resample(freqs[i]).last() # Resample price data
    log_ret = np.log(resampled / resampled.shift(1)) # Calculate log returns
    ann_mean.append(log_ret.mean() * periods[i]) # Annualized mean
    ann_std.append(log_ret.std() * np.sqrt(periods[i])) # Annualized std
```

```python
ann_mean
```

```python
ann_std
```

```python
# Generate summary table
summary = pd.DataFrame(data = {"ann_std": ann_std, "ann_mean":ann_mean}, index=freqs)
summary
```

```python
summary.plot(kind = "scatter", x="ann_std", y="ann_mean", figsize=(15,12), s=50, fontsize=15)
for i in summary.index:
    plt.annotate(i, xy=(summary.loc[i, "ann_std"]+0.0005, summary.loc[i, "ann_mean"]+0.0005), size=15)
plt.ylim(0, 0.3)
plt.xlabel("Annualized Volatility (std)", fontsize=15)
plt.ylabel("Annualized Mean Return", fontsize=15)
plt.title("Mean-Variance Analysis at Different Sampling Frequencies", fontsize=20)
plt.show()
```

Smoothing reduces (observed) Risk

Dubious practices:
* Managing (Manipulating) Performance in Performance reportings
* Adjust frequency to investor's holding period -> Volatility is still there
* Comparing assets with different pricing frequency and pricing mechanism -> e.g. real estate with quarterly valuation vs listed stocks (minutely/hourly/daily prices)

Take Home: When comparing instruments, frequency of underlying data must be the same! Don't compare apples and oranges

# Margin Trading & Levered Returns

Definition: Margin trading refers to the practice of using borrowed funds from a broker to trade a financial asset, which forms the collateral for the loan from the broker

In simple words: Investors don't pay the full price but they get the full benefit (less borrowing cost)

It's a two edged sword: Leverage amplifies both gains and losses. In the event of a loss, the collateral gets reduced and the investor either posts additional margin or the brokers closes the position.

Example

A Trader buys a stock (stock price: 100) on margin (50%). After one day the price increases to 110. Calculate the unlevered return and levered return

```python
P0 = 100
P1 = 110
leverage = 2
margin = P0 / leverage
```

```python
margin
```

```python
unlev_return = (P1 - P0) / P0
unlev_return
```

```python
lev_return = (P1 - P0) / margin
lev_return
```

```python
lev_return == unlev_return * leverage # This relationship is true for simple returns
```

```python
unlev_return = np.log((P1 - P0) / P0 + 1)
unlev_return
```

```python
lev_return = np.log((P1 - P0) / margin + 1)
lev_return
```

```python
lev_return == unlev_return * leverage # This relationship does not hold for log returns
```

To calculate levered returns, don't multiply leverage with log returns

Hypothesis: For (highly) profitable investment: The more leverage, the better?

```python
import pandas as pd
import numpy as np
pd.options.display.float_format = "{:.4f}".format
import matplotlib.pyplot as plt
plt.style.use("bmh")
```

Take Home:
* With leverage you can lose more than the initial margin (in practice: margin call/ margin closeout)
* Even for (highly) profitable instruments: "The more leverage the better" does not hold
* It's a two edged (non-symmetrical) sword: Leverage amplifies losses more than it amplifies gain
